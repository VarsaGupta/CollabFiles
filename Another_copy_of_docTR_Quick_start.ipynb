```python
%matplotlib inline
import os

# Let's pick the desired backend
# os.environ['USE_TF'] = '1'
os.environ['USE_TORCH'] = '1'

import matplotlib.pyplot as plt

from doctr.io import DocumentFile
from doctr.models import ocr_predictor
```

```python
# code

# Instantiate a pretrained model
model = ocr_predictor(pretrained=True)
```

For the next steps, we will need a proper PDF document that will be used to showcase the library features

```python
# code
document = DocumentFile.from_images("/content/test6.jpeg")
```

```python
# code
result = model(document)
```

```python
# code
result.show()
```

```python
# code
# JSON export
json_export = result.export()
print(json_export)
```

```python
json_export["pages"][0]["blocks"][0]["lines"][0]["words"][0]["value"]
```

```python
for words in json_export["pages"][0]["blocks"][0]["lines"][0]["words"]:
    print(words["value"])
```

```python
#  Iterate over pages
for page in json_export.get('pages', []):
    # Iterate over blocks
    for block in page.get('blocks', []):
        # Iterate over lines
        for line in block.get('lines', []):
            # Iterate over words
            for word in line.get('words', []):
                # Extract the value of each word
                word_value = word.get('value', '')
                print(word_value)
```

docTR is, under the hood, running Deep Learning models to perform the different tasks it supports. All of this is wrapper in a `Predictor` object, which will take care of pre-processing, model inference and post-processing for you ⚡

```python
import requests
import json

# Nanonets API endpoint
url = "https://app.nanonets.com/api/v2/OCR/Model/767f75a4-5029-11ee-af13-d6549852b32f/LabelFile"

# Path to the image file
image_path = "/content/pres01.png"  # Replace with the actual path to your image file

# API key
api_key = "767f75a4-5029-11ee-af13-d6549852b32f"

# Prepare headers with API key
headers = {
    "accept": "application/json",
    "Authorization": "Bearer " + api_key
}

# Prepare the payload
payload = {'file': open(image_path, 'rb')}

# Send the POST request to the Nanonets API
response = requests.post(url, files=payload, headers=headers)

# Check if request was successful
if response.status_code == 200:
    # Parse the JSON response
    result = json.loads(response.text)
    # Print the OCR result
    print("OCR Result:")
    for prediction in result["result"]:
        print(prediction["label"], ":", prediction["ocr_text"])
else:
    print("Error:", response.text)
```
